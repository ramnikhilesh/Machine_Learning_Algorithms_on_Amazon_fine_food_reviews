# SGD own implementation

```python
def manual_pred(X,w,b):
    y_hat = []
    for x in X:
        y_hat.append(np.dot(w, x.T) + b)
    return y_hat
```

```python
def grad(w,X,Y,b):
    i = np.random.randint(0, Y.shape[0], 10)
    xb = X[i, :]
    yb = Y[i]
    W_grad = np.zeros(X.shape[1]).reshape(1, X.shape[1])
    n = float(yb.shape[0])
    b_grad = 0
    for k in range(yb.shape[0]):
        W_grad += -(2/n)*(xb[k])*(yb[k] -(np.dot(w, xb[k].T) + b))
        b_grad += -(2/n)*(yb[k] - (np.dot(w, xb[k].T) + b))
    return W_grad, b_grad
```

```python
def SGDmanual(X,Y,r,p):
    w = np.random.normal(0,1,X.shape[1]).reshape(1, X.shape[1])
    b = 0
    mse = []
    for i in range(1, p+1):
        w_gradient, b_gradient = grad(w, X, Y, b)
        w -= r*w_gradient
        b -= r*b_gradient
        y_hat = manual_pred(X, w, b)
        err = mean_squared_error(Y, y_hat)
        mse.append(err)
        r = 0.01/pow(i, 0.25)
    return w, b, mse
```

```python
W,b,mse= SGDmanual(X,Y,0.01,1000)
y_hat= manual_pred(X,W,b)
mse_manual = mean_squared_error(Y, y_hat)
print("The implementation of SGD gives a mse= {}".format(mse_manual))
```
